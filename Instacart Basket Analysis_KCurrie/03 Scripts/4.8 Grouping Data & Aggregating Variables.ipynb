## Low & High Spender Tags

```python
# Create a Spender Range column using transformation aggregation that gathers average price spent by customer

ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check output for specific columns

ords_prods_merge[['user_id', 'prices', 'mean_price']].head()
```

```python
# Create flag for 'High spender'

ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
# Create flag for 'Low spender'

ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
# Check frequency of spender flags

ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

## Customer Frequency Tag

```python
# Create a Median Order column using transformation aggregation that gathers median order made by customer

ords_prods_merge['median_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check output for specific columns

ords_prods_merge[['user_id', 'days_since_prior_order', 'median_order']].head()
```

```python
# Create flag for 'Non-frequent customer'

ords_prods_merge.loc[ords_prods_merge['median_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# Create flag for 'Regular customer'

ords_prods_merge.loc[(ords_prods_merge['median_order'] > 10) & (ords_prods_merge['median_order'] <=20), 'order_frequency_flag'] = 'Regular customer'
```

```python
# Create flag for 'Frequent customer'

ords_prods_merge.loc[ords_prods_merge['median_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Check frequency of order frequency flags

ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Check output for specific columns

ords_prods_merge[['user_id', 'days_since_prior_order', 'median_order', 'order_frequency_flag']].head()
```

```python
print(ords_prods_merge[ords_prods_merge['order_frequency_flag'] == 'nan'][['user_id', 'days_since_prior_order', 'median_order', 'order_frequency_flag']].head())
```
