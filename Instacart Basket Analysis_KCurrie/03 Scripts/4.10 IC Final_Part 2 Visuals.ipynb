# 03. Key Visuals

### Busiest Days & Hours:
The sales team needs to know what the busiest days of the week and hours of the day are (i.e., the days and times with the most orders) in order to schedule ads at times when there are fewer orders.

```python
# Show frequency of orders_day_of_week, sorted

orders_day_of_week_bar = ic_final['orders_day_of_week'].value_counts().sort_index().plot.bar(color=['#FF7009'])
orders_day_of_week_bar.set_facecolor('white')
plt.title('Orders by Day of Week')
plt.xlabel('')
plt.ylabel('Number of orders - in millions')
# Custom x-axis labels
days_custom_labels = ['Sat', 'Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri']  
# Define  custom labels
orders_day_of_week_bar.set_xticklabels(days_custom_labels)
# Disable scientific notation on the y-axis
plt.ticklabel_format(style='plain', axis='y')
# Format y-axis ticks with commas
days_current_values = orders_day_of_week_bar.get_yticks()  # Get current y-tick values
orders_day_of_week_bar.set_yticklabels(['{:,}'.format(int(x)) for x in days_current_values])  # Format with commas
# Adjust layout to prevent clipping of labels
plt.tight_layout()

plt.show()
```

```python
# Show frequency of order_hour_of_day, sorted

orders_hr_of_day_bar = ic_final['order_hour_of_day'].value_counts().sort_index().plot.bar(color=['#003D29'])
orders_hr_of_day_bar.set_facecolor('white')
plt.title('Orders by Hour of Day')
plt.xlabel('')
plt.ylabel('Number of orders - in millions')
# Custom x-axis labels
hrs_custom_labels = ['12am', '1am', '2am', '3am', '4am', '5am', '6am', '7am', '8am', '9am', '10am','11am', '12pm', '1pm', '2pm', '3pm','4pm', '5pm', '6pm','7pm', '8pm', '9pm', '10pm', '11pm'] 
# Define  custom labels
orders_hr_of_day_bar.set_xticklabels(hrs_custom_labels)
# Disable scientific notation on the y-axis
plt.ticklabel_format(style='plain', axis='y')
# Format y-axis ticks with commas
hrs_current_values = orders_day_of_week_bar.get_yticks()  
# Get current y-tick values
orders_hr_of_day_bar.set_yticklabels(['{:,}'.format(int(x)) for x in hrs_current_values])  # Format with commas
# Adjust layout to prevent clipping of labels
plt.tight_layout()

plt.show()
```

```python
# Export outside Jupyter in the form of an image file

orders_hr_of_day_bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'orders_hr_of_day_bar.png'))
```

### Product Popularity:
Are there certain types of products that are more popular than others? The marketing and sales teams want to know which departments have the highest frequency of product orders.

```python
# Show frequency of purchases by department (horizontal)

# Set figure size to make it wider
plt.figure(figsize=(10, 6)) 

departments_frequency_bar = ic_final['department'].value_counts().plot.barh(color=['#0AAD0A'])
departments_frequency_bar.set_facecolor('white')
plt.title('Purchases by Department')
plt.xlabel('Number of Products Purchased')
plt.ylabel('')

# Disable scientific notation on the x-axis
plt.ticklabel_format(style='plain', axis='x')

# Format x-axis ticks with commas
x_current_values = departments_frequency_bar.get_xticks()  
# Get current x-tick values
departments_frequency_bar.set_xticklabels(['{:,}'.format(int(x)) for x in x_current_values]) 

# Reverse the order of the y-axis
departments_frequency_bar.invert_yaxis()

# Adjust layout to prevent clipping of labels
plt.tight_layout()

plt.show()
```

```python
# Export outside Jupyter in the form of an image file

departments_frequency_bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'departments_frequency_bar.png'))
```

### Custom Loyalty & Product Type:
Are there differences in ordering habits based on a customer’s loyalty status?

```python
# Create crosstab of 'loyalty_flag' and 'department'

loyalty_department_crosstab = pd.crosstab(ic_final['department'], ic_final['loyalty_flag'], dropna = True)
```

```python
# Create grouped bar plot of customer loyalty and departments purchased from

# Create a new figure with specific size
fig, ax = plt.subplots(figsize=(20, 12)) 

# Define a list of colors for each loyalty flag
loyalty_colors = ['#FF7009', '#003D29', '#0AAD0A']

# Plot directly on the axes
loyalty_department_crosstab_bar = loyalty_department_crosstab.plot(kind='barh', color=loyalty_colors, ax=ax)
ax.set_facecolor('white')
ax.set_title('Customer Loyalty Purchases by Department')
ax.set_xlabel('Number of Products Purchased')
ax.set_ylabel('')

# Create a custom formatter function
def format_with_commas(x, p):
    return f"{x:,.0f}"

# Apply the custom formatter to the x-axis
ax.xaxis.set_major_formatter(FuncFormatter(format_with_commas))

# Add legend with corrected title
ax.legend(title='Loyalty Flag', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout with some padding
plt.tight_layout(pad=2.0)

# Show the plot
plt.show()
```

```python
# Export outside Jupyter in the form of an image file

loyalty_department_crosstab_bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'loyalty_department_crosstab_bar.png'))
```

### Ordering Habits by Region:
Are there differences in ordering habits based on a customer’s region?

```python
# Create crosstab of 'department' and 'region'

region_department_crosstab = pd.crosstab(ic_final['department'], ic_final['region'], dropna = True)
```

```python
# Create grouped bar plot showing regional department sales

# Create a new figure with specific size
fig, ax = plt.subplots(figsize=(20, 12)) 

# Define a list of colors for each loyalty flag
region_colors = ['#FF7009', '#003D29', '#0AAD0A', '#BA0239']

# Plot directly on the axes
region_department_crosstab_bar = region_department_crosstab.plot(kind='barh', color=region_colors, ax=ax)
ax.set_facecolor('white')
ax.set_title('Ordering Habits by Region')
ax.set_xlabel('Number of Products Purchased')
ax.set_ylabel('')

# Create a custom formatter function
def format_with_commas(x, p):
    return f"{x:,.0f}"

# Apply the custom formatter to the x-axis
ax.xaxis.set_major_formatter(FuncFormatter(format_with_commas))

# Add legend with corrected title
ax.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```
